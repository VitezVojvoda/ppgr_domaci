```python
import numpy as np
import numpy.linalg as la

O3 = np.array([0,0,0])
def dveJed(Ah, A1h):
    jed1 = np.concatenate((O3, -A1h[2]*Ah, A1h[1]*Ah))    # pogledajte slajdove pa popunite
    jed2 = np.concatenate(( A1h[2]*Ah, O3, -A1h[0]*Ah))
    return np.array([jed1, jed2])

def DLT(pts, pts1):
    A = []
    for i in range(len(pts)):
        dve = dveJed(pts[i], pts1[i])
        A.append(dve[0])
        A.append(dve[1])
    np.array(A)
    U,S, Vh = la.svd(A)   # ispraviti
    matP = Vh[-1].reshape(3,3)  #ispraviti
    return matP/matP[2,2]  # normiramo matricu da u donjem desnom uglu bude broj 1
```
