```python
import numpy as np
```

```python
def aff2hom(t):
    return np.append(t, 1)
```

```python
def hom2aff(t):
    if(t[-1] == 0):
        raise ZeroDivisionError("deljenje nulom")
    return t[:-1]/t[-1]
```

```python
def prosek(a,b,c):
    a1 = hom2aff(a)
    b1 = hom2aff(b)
    c1 = hom2aff(c)
    return (a1+b1+c1)/3
```

```python
t1 = np.array([800.0, 425.0, 1.0])
t2 = np.array([365.0, 688.0, 1.0])
t3 = np.array([208.0, 502.0, 1.0])
t5 = np.array([832.0, 283.0, 1.0])
t6 = np.array([350.0, 488.0, 1.0])
t7 = np.array([171.0, 327.0, 1.0])
t8 = np.array([624.0, 212.0, 1.0])
```

```python
def osmoTeme():
    p15 = np.cross(t1, t5)
    p26 = np.cross(t2, t6)
    p37 = np.cross(t3, t7)

    Xb1 = np.cross(p15, p26)
    Xb2 = np.cross(p15, p37)
    Xb3 = np.cross(p26, p37)
    Xb = aff2hom(prosek(Xb1, Xb2, Xb3))

    p12 = np.cross(t1, t2)
    p56 = np.cross(t5, t6)
    p78 = np.cross(t7, t8)

    Yb1 = np.cross(p12, p56)
    Yb2 = np.cross(p12, p78)
    Yb3 = np.cross(p56, p78)
    Yb = aff2hom(prosek(Yb1, Yb2, Yb3))

    p23 = np.cross(t2, t3)
    p58 = np.cross(t5, t8)
    p67 = np.cross(t6, t7)

    Zb1 = np.cross(p23, p58)
    Zb2 = np.cross(p23, p67)
    Zb3 = np.cross(p58, p67)
    Zb = aff2hom(prosek(Zb1, Zb2, Zb3))

    p48 = np.cross(Xb, t8)
    p34 = np.cross(Yb, t3)
    p14 = np.cross(Zb, t1)

    t41 = np.cross(p48, p34)
    t42 = np.cross(p48, p14)
    t43 = np.cross(p34, p14)
    t4 = prosek(t41, t42, t43)
    return t4
```

```python
osmoTeme()
```
